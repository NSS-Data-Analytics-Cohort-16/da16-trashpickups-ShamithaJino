```python
df_trash = pd.read_csv('../data/trash_hauler_report_with_lat_lng.csv')
df_trash.head()
```

```python
df_trash['Trash Hauler'].unique()
```

```python
# Filters only 'RED RIVER' Trash Hauler
df_red_river = df_trash[df_trash['Trash Hauler'] == 'RED RIVER']
df_red_river.head()
```

```python
df_trash['Request'].value_counts()
```

```python
df_missed_pickup = df_red_river[df_red_river['Request'].isin(['Trash - Curbside/Alley Missed Pickup', 'Trash - Backdoor', 'Trash Collection Complaint'])]
df_missed_pickup.head()
```

```python
df_value_count = df_missed_pickup[['Trash Hauler','Incident Address']].value_counts().reset_index()
df_value_count.head()
```

```python
df_missed_2plus = df_value_count[df_value_count['count']>1]
df_missed_2plus['fine']= (df_missed_2plus['count']-1) *200
#df_missed_2plus.head()
df_missed_2plus.groupby('Trash Hauler')['fine'].sum()
```

### ⦁	How do metro crews compare to the contractor's performance?

```python
df_trash.groupby('Trash Hauler')['Request Number'].count()
```

```python
df_summary = df_trash.groupby('Trash Hauler').agg(total_request= ('Request Number', 'count'),total_routes=('Trash Route','nunique') )
df_summary['request_per_route'] = (df_summary['total_request']/df_summary['total_routes']).round(decimals = 0).astype(int)
df_summary.head()
```

### ⦁	How much does each trash hauler owe?


```python
df_hauler_fines = df_trash[df_trash['Request'].isin(['Trash - Curbside/Alley Missed Pickup', 'Trash - Backdoor', 'Trash Collection Complaint'])]
df_address_count = df_hauler_fines.groupby(['Trash Hauler','Incident Address']).agg(total_count= ('Request Number','count')).reset_index()
df_address_count_2plus = df_address_count[df_address_count['total_count']>1]
df_address_count_2plus['total_fine']=(df_address_count['total_count']-1) *200
df_address_grouped= df_address_count_2plus.groupby('Trash Hauler')['total_fine'].sum().reset_index()
df_address_grouped.head()
 
```
